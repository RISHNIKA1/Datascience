```python
# Example of encapsulation

class BankAccount :
    def __init__(self,name,balance):
        self.name = name                # public attribute
        self.__balance = balance        # private attribute
        
    def deposit(self,amount):
        if amount >0:
            self.__balance += amount
    
    def get_balance(self):
        return self.__balance
    

account = BankAccount("Rishi",25000)
account.deposit(5000)

print(account.name)                # Allowed
print(account.get_balance())       # Allow this method
#print(account.__balance)           # Error cant access private variable directly
        
```

```python
# Example of Abstract classes

from abc import ABC,abstractmethod

# Abstract Class
class Vehicle(ABC):
    
    @abstractmethod
    def start_engine(self):
        pass
    
    @abstractmethod
    def stop_engine(self):
        pass
    
# child class
class car(Vehicle):
    def start_engine(self):
        print("car engine started")
    
    def stop_engine(self):
        print("car engine stopped")
        

# creating object of abstract class will raise error
# V= Vehicle() # wrong

# but we can use child class

c=car()
c.start_engine()
c.stop_engine()
    
    
```

```python
# Inheriatnce 

class Parent:
    def display(self):
        print("hi")
    
class child(Parent):
    pass

c = child()
c.display() #inherit method
```

```python
# Example of Single Inheritace

class Animal:
    def sound(self):
        print("Animal sound")
    
class dog(Animal):
    pass

d= dog()
d.sound()
        
```

```python
# Example of Multiple Inheritance

class Father():
    def Money(self):
        print("dads money")
class Mother():
    def Phone(self):
        print("Moms Phone")
class Child(Father,Mother):
    pass

son = Child()
son.Money()
son.Phone()

```

```python
# Examole of Multilevel Inheritance

class Grandparent:
    def family(self):
        print("Grandparent family")

class Father(Grandparent):
    pass

class son(Father):
    pass

child = son()
child.family()
```

```python
# Example of Hireachical inheritance

class vehicle:
    def fuel(self):
        print("Petrol or Diesel")
    
class car(vehicle):
    pass

class Bike(vehicle):
    pass

c=car()
b= Bike()
c.fuel()
b.fuel()
```

```python
# Example of Hybrid Inheritance

class A:
    def a(self):
        print("A")
class B(A):
    def b(self):
        print("B")
class C:
    def c(self):
        print("C")
class D(B,C):
    def d(self):
        print("D")

obj=D()
obj.a()
obj.b()
obj.c()
obj.d()
```

```python
# Example Method Overridng

class Father:
    def house(self):
        print("White house")

class Son(Father):
    def phone(self):
        print("sons phone")
        
    def house(self):
        print("yellow house")
        
s=Son()
s.house()
```
